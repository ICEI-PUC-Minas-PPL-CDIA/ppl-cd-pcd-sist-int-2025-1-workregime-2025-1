### Transformando dados relacionados a segurança por região

```python
# Somando os roubos em cada linha
data['Total de roubos'] = data[['Roubos de veículo', 'Roubos de carro', 'Roubos de moto', 'Roubos fora do domicílio']].sum(axis=1)
data['Total de roubos'].value_counts(dropna=False)

descriptive_labels = ['Baixo', 'Moderado', 'Alto', 'Muito alto']

_, bins_edges = pd.qcut(data['Total de roubos'], q=4, duplicates='drop', retbins=True)

num_actual_bins = len(bins_edges) - 1

if num_actual_bins > len(descriptive_labels):
    print(f"Warning: Not enough descriptive labels for {num_actual_bins} bins.")
    labels = descriptive_labels + [f'Bin {i+len(descriptive_labels)}' for i in range(num_actual_bins - len(descriptive_labels))]
else:
    labels = descriptive_labels[:num_actual_bins]

data['Nivel de segurança'] = pd.cut(data['Total de roubos'], bins=bins_edges, labels=labels, include_lowest=True)

# Exibindo o resultado
data['Nivel de segurança'].value_counts(dropna=False)
```
