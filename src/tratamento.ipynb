```python
data = pd.read_csv('state_of_data_enriquecida.csv')
data.head()
```

```python
data.rename(columns={"('P1_a ', 'Idade')": "Idade",
                     "('P1_b ', 'Genero')": "Genero",
                     "('P1_c ', 'Cor/raca/etnia')": "Cor/Raca",
                     "Média de horas com trabalho doméstico e cuidado de pessoas": "Trabalho domestico",
                     "('P1_d ', 'PCD')": "PCD",
                     "('P1_i ', 'Estado onde mora')": "Estado onde mora",
                     "('P1_i_2 ', 'Regiao onde mora')": "Região onde mora",
                     "Roubos fora do domicílio por região": "Roubos fora do domicilio",
                     "('P1_l ', 'Nivel de Ensino')": "Nivel de ensino",
                     "('P1_m ', 'Área de Formação')": "Area de formação",
                     "('P2_a ', 'Qual sua situação atual de trabalho?')": "Situação atual de trabalho",
                     "('P2_f ', 'Cargo Atual')": "Cargo atual",
                     "('P2_g ', 'Nivel')": "Nivel",
                     "('P2_h ', 'Faixa salarial')": "Faixa salarial",
                     "('P2_i ', 'Quanto tempo de experiência na área de dados você tem?')": "Tempo de experiencia",
                     "('P2_r ', 'Atualmente qual a sua forma de trabalho?')": "Forma de trabalho atual",
                     "('P2_s ', 'Qual a forma de trabalho ideal para você?')": "Forma de trabalho ideal"}, inplace=True)
data.head()
```

```python
# Retirando valores NA na coluna-classe
data = data.dropna(subset=['Forma de trabalho ideal'])
data.head()
```

```python
data.isna().sum()
```

```python
data['Estado onde mora'].value_counts(dropna=False)
```

```python
data['Estado onde mora'].fillna('São Paulo (SP)', inplace=True)
```

```python
data['Região onde mora'].value_counts(dropna=False)
```

```python
data['Região onde mora'].fillna('Sudeste', inplace=True)
```

```python
data.isna().sum()
```

```python
qtd = data[data['Região onde mora'] == 'Sudeste']['Roubos de veículo'].values[0]
data['Roubos de veículo'].fillna(qtd, inplace=True)
qtd = data[data['Região onde mora'] == 'Sudeste']['Roubos de carro'].values[0]
data['Roubos de carro'].fillna(qtd, inplace=True)
qtd = data[data['Região onde mora'] == 'Sudeste']['Roubos de moto'].values[0]
data['Roubos de moto'].fillna(qtd, inplace=True)
qtd = data[data['Região onde mora'] == 'Sudeste']['Roubos de bicicleta'].values[0]
data['Roubos de bicicleta'].fillna(qtd, inplace=True)
qtd = data[data['Região onde mora'] == 'Sudeste']['Roubos fora do domicilio'].values[0]
data['Roubos fora do domicilio'].fillna(qtd, inplace=True)
```

```python
data.isna().sum()
```

```python
data.fillna('Não informado', inplace=True)
```

```python
data.to_csv('data.csv')
```
